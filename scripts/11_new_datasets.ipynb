```python
from sklearn.semi_supervised import LabelSpreading
import pickle, glob
from skimage.io import imread
import numpy as np, glob, tqdm


d=pd.read_pickle("images/all_leukocytes_arrlist.p")
d2=pd.read_pickle("images/new_leuks.p")
d3=pd.read_pickle("images/new_leuks_cluster.p")
coords=np.array(pd.Series(d2).map(lambda x: x.shape[:2]).tolist())
coords2=np.array(pd.Series(d3).map(lambda x: x.shape[:2]).tolist())
coords3=np.array(pd.Series(d).map(lambda x: x.shape[:2]).tolist())
lp=LabelSpreading(kernel="knn",n_neighbors=20,alpha=0.7).fit(np.log(np.vstack([coords,coords2,coords3])),[0]*len(coords)+[1]*len(coords2)+[-1]*len(coords3))
is_leuk_cluster=lp.transduction_[-len(coords3):]==1
```

```python
arrs={f.split('_')[-2]:pd.Series(pd.read_pickle(f)) for f in glob.glob('images/all_*_arrlist.p')}

arrs['leuk_cluster']=arrs['leukocytes'].loc[is_leuk_cluster]#.tolist()
arrs['leukocytes']=arrs['leukocytes'].loc[~is_leuk_cluster]#.tolist()
arrs={k:v.tolist() for k,v in arrs.items()}
new_imgs={}
for k in ["aty","neg","squamous","leuk_clusters_deident"]:
    k2=k
    if k.startswith("leuk_clusters"): k2="leuk_cluster"
    new_dat=[imread(img) for img in tqdm.tqdm(glob.glob(f"images/{k}/*.png"))]
    print(k2,len(new_dat),len(arrs[k2]))
    arrs[k2].extend(new_dat)
    
arrs['aty'].extend(pd.read_pickle("images/new_aty.p"))
arrs['leukocytes'].extend(pd.read_pickle("images/new_leuks.p"))
arrs['leuk_cluster'].extend(pd.read_pickle("images/new_leuks_cluster.p"))
arrs['blue_granular']=pd.read_pickle("images/blue_granular.p")
```

```python
cell_types=list(arrs.keys())
cell_types.remove("unsorted")
```

```python
import numpy as np, pandas as pd
X=[]
y=[]
for k in cell_types:
    print(k)
    X.extend(arrs[k])
    y.extend([k]*len(arrs[k]))
y=np.array(y)
X=pd.Series(X)
```

```python
from sklearn.preprocessing import LabelEncoder
LabelEncoder().fit("""leukocytes
neg
squamous
aty
rbc
debris
crystal
leuk_cluster
blue_granular""".splitlines()).classes_
```

```python
import cv2
from skimage import measure
from sklearn.preprocessing import QuantileTransformer,StandardScaler, RobustScaler
from tqdm import tqdm, trange
#

stats = X.map(lambda x: measure.regionprops((cv2.cvtColor(x, cv2.COLOR_RGB2GRAY)<255).astype(int), coordinates='rc')[0]).tolist()
stats_dict = {num:{} for num in range(len(stats))}
for r in trange(len(stats_dict)):
    for key in stats[r]:
        stats_dict[r][key] = stats[r][key]
        
```

```python
df_stat=pd.DataFrame([{k:v for k,v in stats_dict[i].items() if isinstance(v,(np.int64,int,float,tuple)) and k not in ['slice','local_centroid']} for i in range(len(X))])
df_stat=df_stat.drop(columns=['bbox','bbox_area','centroid','euler_number','label','orientation','perimeter_crofton'])
```
